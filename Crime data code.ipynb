### 1.Loading the dataset

```python
import pandas as pd

# Load the dataset
file_path = 'Crime_Data_from_2020_to_Present.csv'
crime_data = pd.read_csv(file_path)

# Display the column names to identify the correct ones
print("Column names in the dataset:")
print(crime_data.columns)

# Handle missing data and duplicates as earlier described
crime_data_cleaned = crime_data.dropna().drop_duplicates()

# Example of converting a date column, replace 'Date_Column_Name' with the actual column name
# crime_data_cleaned['Date_Column_Name'] = pd.to_datetime(crime_data_cleaned['Date_Column_Name'], errors='coerce')

# Identify and print categorical columns
categorical_columns = crime_data_cleaned.select_dtypes(include=['object']).columns
print("\nCategorical columns identified for encoding:")
print(categorical_columns)

# Apply one-hot encoding to the identified categorical columns
crime_data_cleaned = pd.get_dummies(crime_data_cleaned, columns=categorical_columns, drop_first=True)

# Display the cleaned data
print("\nCleaned Data:")
print(crime_data_cleaned.head())
```

### 2.Checking for the missing data

```python
# Checking for the missing values
missing_data = crime_data.isnull().sum()
print("Missing Data:\n", missing_data)
```

### 3.Removing Missing Values

```python
# Droping the columns with more than 30% missing data
crime_data_cleaned = crime_data.dropna(thresh=len(crime_data) * 0.7, axis=1)

# Droping the rows with any remaining missing data
crime_data_cleaned = crime_data_cleaned.dropna()

# Checking if the missing values have been removed
print(crime_data_cleaned.isnull().sum())
```

### 4.Removing the duplicate rows

```python
# Removing the duplicate rows
crime_data_cleaned = crime_data_cleaned.drop_duplicates()

# Checking if the duplicates have been removed
print("Number of duplicate rows after cleaning:", crime_data_cleaned.duplicated().sum())
```

### 5.Checking for the column names

```python
# Checking the column names
print(crime_data_cleaned.columns)
```

### 6.Converting data types

```python
# Converting 'DATE OCC' column to datetime
crime_data_cleaned['DATE OCC'] = pd.to_datetime(crime_data_cleaned['DATE OCC'], errors='coerce')

# Checkich the data types to confirm the conversion
print(crime_data_cleaned.dtypes)

# Checking the format of the 'DATE OCC' column
print(crime_data_cleaned['DATE OCC'].head())
```

### 7.Handling Outliers

#### Outliers in 'Vict Age'

```python
# Identifying and removing outliers using the IQR method for 'Vict Age'
Q1 = crime_data_cleaned['Vict Age'].quantile(0.25)
Q3 = crime_data_cleaned['Vict Age'].quantile(0.75)
IQR = Q3 - Q1

# Filtering out the outliers
crime_data_cleaned = crime_data_cleaned[~((crime_data_cleaned['Vict Age'] < (Q1 - 1.5 * IQR)) | (crime_data_cleaned['Vict Age'] > (Q3 + 1.5 * IQR)))]

# Checking the dataset after outlier removal
print(crime_data_cleaned.describe())
```

#### Outliers in 'LAT' and 'LON'

```python
# Identifying and removing outliers for 'LAT' and 'LON' using IQR
Q1_lat = crime_data_cleaned['LAT'].quantile(0.25)
Q3_lat = crime_data_cleaned['LAT'].quantile(0.75)
IQR_lat = Q3_lat - Q1_lat

Q1_lon = crime_data_cleaned['LON'].quantile(0.25)
Q3_lon = crime_data_cleaned['LON'].quantile(0.75)
IQR_lon = Q3_lon - Q1_lon

# Filtering out geographical outliers for 'LAT'
crime_data_cleaned = crime_data_cleaned[~((crime_data_cleaned['LAT'] < (Q1_lat - 1.5 * IQR_lat)) | (crime_data_cleaned['LAT'] > (Q3_lat + 1.5 * IQR_lat)))]

# Filtering out geographical outliers for 'LON'
crime_data_cleaned = crime_data_cleaned[~((crime_data_cleaned['LON'] < (Q1_lon - 1.5 * IQR_lon)) | (crime_data_cleaned['LON'] > (Q3_lon + 1.5 * IQR_lon)))]

# Checking the dataset after outlier removal
print(crime_data_cleaned[['LAT', 'LON']].describe())
```

#### Outliers in 'Premis Cd'

```python
# Identifying and removing outliers using the IQR method for 'Premis Cd'
Q1 = crime_data_cleaned['Premis Cd'].quantile(0.25)
Q3 = crime_data_cleaned['Premis Cd'].quantile(0.75)
IQR = Q3 - Q1

# Filtering out the outliers
crime_data_cleaned = crime_data_cleaned[~((crime_data_cleaned['Premis Cd'] < (Q1 - 1.5 * IQR)) | (crime_data_cleaned['Premis Cd'] > (Q3 + 1.5 * IQR)))]

# Checking the dataset after outlier removal
print(crime_data_cleaned['Premis Cd'].describe())
```

### 8.Standardize/Normalize the Numerical Data

```python
from sklearn.preprocessing import StandardScaler

# Selecting numerical columns to standardize
numerical_columns = ['Vict Age', 'LAT', 'LON']  # Replace with actual column names

# Initializing StandardScaler
scaler = StandardScaler()

# Applying standardization
crime_data_cleaned[numerical_columns] = scaler.fit_transform(crime_data_cleaned[numerical_columns])

# Checking the transformed data
print(crime_data_cleaned[numerical_columns].head())
```

### 9.Encoding the Categorical Data

```python
from sklearn.preprocessing import LabelEncoder

# Selecting categorical columns
categorical_columns = ['Crm Cd Desc', 'LOCATION']  # Replace 'Crime Type' with 'Crm Cd Desc'

# Applying label encoding
encoder = LabelEncoder()
for col in categorical_columns:
    crime_data_cleaned[col] = encoder.fit_transform(crime_data_cleaned[col])

# Checking the encoded data
print(crime_data_cleaned[categorical_columns].head())
```

### 10.Final Cleaned data for further analysis

# 5.ADVANCED ANALYSIS USING SARIMA

### SARIMA - Seasonal Autoregressive Integrated Moving Average
#### 1) It is a versatile and widely used time series forecasting model. 
#### 2) It's an extension of the non-seasonal ARIMA model, designed to handle data with seasonal patterns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from pandas.plotting import autocorrelation_plot

# Load and preprocess the dataset
data = pd.read_csv('Crime_Data_from_2020_to_Present.csv')  # Replace with the actual path to your data

# Strip the time part and only keep the date (if you want only the date)
data['DATE OCC'] = data['DATE OCC'].str.split(' ').str[0]
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], format='%m/%d/%Y')

# Group the data by year and month to get monthly crime counts
data['YearMonth'] = data['DATE OCC'].dt.to_period('M')  # Group by year and month
monthly_crime_count = data.groupby('YearMonth').size().reset_index(name='Crime_Count')

# Convert YearMonth back to datetime for proper plotting
monthly_crime_count['YearMonth'] = monthly_crime_count['YearMonth'].dt.to_timestamp()

# Log-transform the data (add 1 to avoid log(0))
monthly_crime_count['Log_Crime_Count'] = np.log1p(monthly_crime_count['Crime_Count'])

# Plot the original log-transformed crime count data
plt.figure(figsize=(10,6))
plt.plot(monthly_crime_count['YearMonth'], monthly_crime_count['Log_Crime_Count'])
plt.title('Log-Transformed Crime Counts Over Time')
plt.xlabel('Date')
plt.ylabel('Log Crime Count')
plt.xticks(rotation=45)
plt.show()

# Check for stationarity using Augmented Dickey-Fuller test
def adf_test(series):
    result = adfuller(series)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    if result[1] > 0.05:
        print('Non-Stationary')
    else:
        print('Stationary')

adf_test(monthly_crime_count['Log_Crime_Count'])

# Fit SARIMA model with (p, d, q) and seasonal (P, D, Q, S) parameters
order = (1, 1, 1)  # Adjust based on ACF and PACF plots
seasonal_order = (1, 1, 0, 12)  # Seasonal order (P, D, Q, S)

# Build and fit the SARIMA model
sarima_model = SARIMAX(monthly_crime_count['Log_Crime_Count'], order=order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
sarima_fit = sarima_model.fit(disp=False)

# Model summary
print(sarima_fit.summary())

# Forecasting for the next 12 months
forecast_periods = 12
forecast = sarima_fit.get_forecast(steps=forecast_periods)
forecast_values = forecast.predicted_mean
forecast_conf_int = forecast.conf_int()

# Convert the forecast back to the original scale (exponentiate)
forecast_values_exp = np.expm1(forecast_values)  # inverse of log1p

# Set any negative forecasts to zero (as crime counts cannot be negative)
forecast_values_exp = np.where(forecast_values_exp < 0, 0, forecast_values_exp)

# Create forecast index starting from the last date in the dataset
forecast_index = pd.date_range(monthly_crime_count['YearMonth'].max(), periods=forecast_periods, freq='MS')

# Plot the original crime count and the forecast
plt.figure(figsize=(10,6))
plt.plot(monthly_crime_count['YearMonth'], monthly_crime_count['Crime_Count'], label='Observed')
plt.plot(forecast_index, forecast_values_exp, color='r', label='Forecast')
plt.fill_between(forecast_index, np.expm1(forecast_conf_int.iloc[:, 0]), np.expm1(forecast_conf_int.iloc[:, 1]), color='pink', alpha=0.3)
plt.title('Crime Count Forecast for Next 12 Months')
plt.xlabel('Date')
plt.ylabel('Crime Count')
plt.legend()
plt.show()

# Print the forecasted crime counts
print("Forecasted Crime Counts for Next 12 Months:")
print(forecast_values_exp)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Crime_Data_from_2020_to_Present.csv')  # Adjust the path as needed

# Step 1: Data Preprocessing and Handling Date Parsing Issues
try:
    data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], format='%m/%d/%Y', errors='coerce')
except:
    data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], errors='coerce')

# Drop rows where 'DATE OCC' could not be converted
data.dropna(subset=['DATE OCC'], inplace=True)

# Aggregate the data by year and month to get the total number of crimes per month
data['YearMonth'] = data['DATE OCC'].dt.to_period('M')  # Create a YearMonth column
monthly_crime_count = data.groupby('YearMonth').size().reset_index(name='Crime_Count')  # Group and count crimes

# Convert 'YearMonth' to datetime for time series analysis
monthly_crime_count['YearMonth'] = monthly_crime_count['YearMonth'].dt.to_timestamp()

# Step 2: Plotting Monthly Crime Counts

# Plotting the monthly crime counts over time
plt.figure(figsize=(10,6))
plt.plot(monthly_crime_count['YearMonth'], monthly_crime_count['Crime_Count'], linestyle='-', marker='o', color='b')
plt.title('Monthly Crime Count Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.grid(True)  # Add grid for better visibility
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Augmented Dickey-Fuller Test
def adf_test(series):
    result = adfuller(series)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    if result[1] > 0.05:
        print('Non-Stationary Data, differencing needed.')
    else:
        print('Stationary Data')

# Run the test
adf_test(monthly_crime_count['Crime_Count'])

# If the data is non-stationary, apply first differencing
monthly_crime_count['Crime_Count_Diff'] = monthly_crime_count['Crime_Count'].diff().dropna()

# Plot differenced data
plt.figure(figsize=(10,6))
plt.plot(monthly_crime_count['YearMonth'][1:], monthly_crime_count['Crime_Count_Diff'], marker='o')
plt.title('Differenced Monthly Crime Count')
plt.xlabel('Date')
plt.ylabel('Differenced Number of Crimes')
plt.xticks(rotation=45)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF and PACF for original and differenced data
plt.figure(figsize=(12,6))

# ACF and PACF of original data
plt.subplot(121)
plot_acf(monthly_crime_count['Crime_Count'], ax=plt.gca(), lags=20)
plt.title('ACF of Original Data')

plt.subplot(122)
plot_pacf(monthly_crime_count['Crime_Count'], ax=plt.gca(), lags=20)
plt.title('PACF of Original Data')
plt.show()

# ACF and PACF of differenced data
plt.figure(figsize=(12,6))

plt.subplot(121)
plot_acf(monthly_crime_count['Crime_Count_Diff'].dropna(), ax=plt.gca(), lags=20)
plt.title('ACF of Differenced Data')

plt.subplot(122)
plot_pacf(monthly_crime_count['Crime_Count_Diff'].dropna(), ax=plt.gca(), lags=20)
plt.title('PACF of Differenced Data')
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Define the model parameters based on the ACF/PACF plots
p = 1  # AR terms
d = 1  # Differencing
q = 1  # MA terms

P = 1  # Seasonal AR
D = 1  # Seasonal Differencing
Q = 0  # Seasonal MA
S = 12  # Seasonality period (12 months for yearly seasonality)

# Build and fit the SARIMA model
sarima_model = SARIMAX(monthly_crime_count['Crime_Count'], 
                       order=(p, d, q), 
                       seasonal_order=(P, D, Q, S),
                       enforce_stationarity=False, 
                       enforce_invertibility=False)

sarima_fit = sarima_model.fit(disp=False)

# Print model summary
print(sarima_fit.summary())
```

```python
# Forecast for the next 12 months
forecast_steps = 12
forecast = sarima_fit.get_forecast(steps=forecast_steps)
forecast_values = forecast.predicted_mean
forecast_conf_int = forecast.conf_int()

# Generate the forecast index (future dates)
last_date = monthly_crime_count['YearMonth'].max()
forecast_index = pd.date_range(last_date, periods=forecast_steps, freq='MS')

# Plot the forecasted values
plt.figure(figsize=(10,6))
plt.plot(monthly_crime_count['YearMonth'], monthly_crime_count['Crime_Count'], label='Observed', marker='o')
plt.plot(forecast_index, forecast_values, color='red', label='Forecast', marker='o')

# Plot the confidence intervals
plt.fill_between(forecast_index, forecast_conf_int.iloc[:, 0], forecast_conf_int.iloc[:, 1], 
                 color='pink', alpha=0.3)

plt.title('Crime Count Forecast for Next 12 Months')
plt.xlabel('Date')
plt.ylabel('Crime Count')
plt.legend()
plt.show()

# Print forecasted crime counts
print("Forecasted Crime Counts for Next 12 Months:")
print(forecast_values)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Load the dataset
data = pd.read_csv('Crime_Data_from_2020_to_Present.csv')  # Adjust the path as needed

# Step 1: Data Preprocessing and Handling Date Parsing Issues
try:
    data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], format='%m/%d/%Y', errors='coerce')
except:
    data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], errors='coerce')

# Drop rows where 'DATE OCC' could not be converted
data.dropna(subset=['DATE OCC'], inplace=True)

# Aggregate the data by year and month to get the total number of crimes per month
data['YearMonth'] = data['DATE OCC'].dt.to_period('M')  # Create a YearMonth column
monthly_crime_count = data.groupby('YearMonth').size().reset_index(name='Crime_Count')  # Group and count crimes

# Convert 'YearMonth' to datetime for time series analysis
monthly_crime_count['YearMonth'] = monthly_crime_count['YearMonth'].dt.to_timestamp()

# Step 2: Check for Stationarity with ADF Test
def adf_test(series):
    if series.isnull().sum() > 0 or len(series) == 0:
        raise ValueError("Series contains missing values or is empty, cannot run ADF test.")
    
    result = adfuller(series)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    if result[1] > 0.05:
        print('Non-Stationary Data, differencing required.')
    else:
        print('Stationary Data')

# Check for missing or empty values before running ADF test
if monthly_crime_count['Crime_Count'].isnull().sum() == 0 and len(monthly_crime_count['Crime_Count']) > 0:
    adf_test(monthly_crime_count['Crime_Count'])
else:
    print("Monthly crime count series contains missing values or is empty.")

# Step 3: Differencing if Necessary
monthly_crime_count['Crime_Count_Diff'] = monthly_crime_count['Crime_Count'].diff().dropna()

# Ensure there are enough data points after differencing
if len(monthly_crime_count['Crime_Count_Diff']) < 2:
    raise ValueError("Not enough data points after differencing for further analysis.")

# Step 4: Plot ACF and PACF to Determine ARIMA(p, d, q) Parameters

plt.figure(figsize=(12,6))

# Plot ACF
plt.subplot(121)
plot_acf(monthly_crime_count['Crime_Count_Diff'], lags=20, ax=plt.gca())
plt.title('ACF of Differenced Data')

# Plot PACF
plt.subplot(122)
plot_pacf(monthly_crime_count['Crime_Count_Diff'], lags=20, ax=plt.gca())
plt.title('PACF of Differenced Data')
plt.show()

# Step 5: Fit ARIMA Model Based on Identified (p, d, q) Parameters

# Skip differencing and fit the ARIMA model directly if the data is small or already stationary
p, d, q = 1, 0, 1  # Assuming no differencing is needed

# Fit the ARIMA model without differencing
arima_model = ARIMA(monthly_crime_count['Crime_Count'], order=(p, d, q))
arima_fit = arima_model.fit()

# Summary of the model
print(arima_fit.summary())


# Step 6: Forecasting

# Forecast for the next 12 months
forecast_steps = 12
forecast = arima_fit.get_forecast(steps=forecast_steps)
forecast_values = forecast.predicted_mean
forecast_conf_int = forecast.conf_int()

# Create a forecast index starting from the last date
forecast_index = pd.date_range(monthly_crime_count['YearMonth'].max(), periods=forecast_steps, freq='MS')

# Plot the forecasted values
plt.figure(figsize=(10,6))
plt.plot(monthly_crime_count['YearMonth'], monthly_crime_count['Crime_Count'], label='Observed', marker='o')
plt.plot(forecast_index, forecast_values, color='red', label='Forecast', marker='o')

# Plot the confidence intervals
plt.fill_between(forecast_index, forecast_conf_int.iloc[:, 0], forecast_conf_int.iloc[:, 1], color='pink', alpha=0.3)

plt.title('Crime Count Forecast for Next 12 Months (ARIMA)')
plt.xlabel('Date')
plt.ylabel('Crime Count')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Print forecasted crime counts
print("Forecasted Crime Counts for Next 12 Months:")
print(forecast_values)
```
